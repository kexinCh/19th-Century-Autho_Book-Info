Read in book_counts csv for word counts in each book

```python
word_count = {}
fnf = pd.read_csv('files_not_found.csv')
fnf['id'] = fnf['filename'].str.split('_').str[0]
# book_filtered = background[~background['id'].isin(fnf['id'])]
book_filtered = sf[~sf['id'].isin(fnf['id'])]

for index, row in book_filtered.iterrows():
    idnum = row['id']
    filename = "counts/csv/" + idnum + "_counts.csv"
    file = pd.read_csv(filename)
    for index, row in file.iterrows():
        word = row[0]
        count = int(row[1])
        if word in word_count:
            word_count[word] += count
        else:
            word_count[word] = count
df = pd.DataFrame(word_count.items(), columns=["word", "count"])
display(df)
```

```python
corpos = df['count'].sum()
# # df['p(w)'] = np.log(df['count'] / (corpos - df['count']))
# # df['corpos_N'] = corpos
# # df = df.rename(columns={'count':'count_n'})
df['corpos_S'] = corpos
df = df.rename(columns={'count':'count_s'})
sf = df
# display(sf)
# Background = df
# display(fantasy)
```

```python
import nltk
from nltk.corpus import names
# nltk.download('names')
common_names = set(names.words())
```

```python
B1800 = pd.read_csv("book1800counts.csv")
B1850 = pd.read_csv("book1850counts.csv")
B1900 = pd.read_csv("book1900counts.csv")
```

```python
B1800sum = B1800['1800count'].sum()
B1800['1800percentage'] = B1800['1800count']/B1800sum
B1850sum = B1850['1850count'].sum()
B1850['1850percentage'] = B1850['1850count']/B1850sum
B1900sum = B1900['1900count'].sum()
B1900['1900percentage'] = B1900['1900count']/B1900sum
display(B1800)
```

```python
words_dif = pd.merge(B1800,B1850,on='word')
words_dif = pd.merge(words_dif, B1900, on='word')
words_dif = words_dif.drop(columns=['Unnamed: 0'])
display(words_dif)
```

```python
words_dif['increase1'] = (words_dif['1850percentage'] - words_dif['1800percentage']) / words_dif['1800percentage']
words_dif['increase2'] = (words_dif['1900percentage'] - words_dif['1850percentage']) / words_dif['1850percentage']
words_dif['total_increase'] = (words_dif['1900percentage'] - words_dif['1800percentage']) / words_dif['1800percentage']
words_dif['biggest_change'] = abs(words_dif['increase1'] * words_dif['increase2'])
# words_dif = words_dif.sort_values(by='increase_appearance_100y', ascending=False)
# words_dif50 = words_dif.sort_values(by='increase_appearance_50y', ascending=False)
display(words_dif)
```

```python
words_dif['word'] = words_dif['word'].astype(str).apply(lambda x: x.title())
words_no_name = words_dif[~words_dif['word'].isin(common_names)]
increase = words_no_name[(words_no_name['increase1'] > 0) & (words_no_name['increase2'] > 0)]
increase = increase.sort_values(by='total_increase', ascending = False)
display(increase.head(50))
```

```python
decrease = words_no_name[(words_no_name['increase1'] < 0) & (words_no_name['increase2'] < 0)]
decrease = decrease.sort_values(by='total_increase', ascending = True)
display(decrease.head(50))
```
