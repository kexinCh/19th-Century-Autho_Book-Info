```python
import pandas as pd
```

```python
from qwikidata.sparql import return_sparql_query_results

# Define your SPARQL query
sparql_query = """
SELECT ?book ?bookLabel ?publicationYLabel ?author ?authorLabel ?genderLabel
WHERE {
  ?book rdfs:label "Harry Potter and the Philosopher's Stone"@en.
  ?book wdt:P31 ?worktype; 
        wdt:P50 ?author.
  OPTIONAL { ?book wdt:P577 ?publicationdate.}
  FILTER(?worktype = wd:Q7725634 || ?worktype = wd:Q1279564 || ?worktype = wd:Q47461344 || ?worktype = wd:Q27217074 || ?worktype = wd:Q732577|| ?worktype = wd:Q8261) #41298 magazine 
  ?author   wdt:P21 ?gender.
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}
"""

print('hek')

# Run the SPARQL query
query_results = return_sparql_query_results(sparql_query)

print(query_results)
# Extract the bindings from the results
bindings = query_results["results"]["bindings"]

# Process the bindings to create a list of dictionaries
result_list = []
for binding in bindings:
    result_dict = {}
    for var, value_info in binding.items():
        result_dict[var] = value_info["value"]
    result_list.append(result_dict)

# Display the list of dictionaries
print(result_list)
```

```python
df = pd.DataFrame(result_list)
display(df)
```
