```python
import sys
import pandas as pd
```

```python
result1 = pd.read_csv('result31290.csv')
result1 = result1.drop_duplicates(subset=['book','id'])
result1['book_QN'] = result1['book'].str.split('/').str[-1]
result1['author_QN'] = result1['author_y'].str.split('/').str[-1]
display(result1)
```

```python
result2 = pd.read_csv('result51287.csv')
result3 = pd.read_csv('result53266.csv')
result4 = pd.read_csv('result59505.csv')
result5 = pd.read_csv('result68194.csv')
result6 = pd.read_csv('result71394.csv')
result7 = pd.read_csv('resultEnd.csv')
```

```python
result = pd.concat([result1,result2,result3,result4,result5,result6,result7])
result = result.drop_duplicates(subset=['book_QN','author_QN','id'])
display(result)
```

```python
author = result['author_x']
f_name = []
m_name = []
for row in author.values:
    if pd.notna(row):
        row = str(row)
        if "," in row:
            parts = row.split(", ")
            name = parts[1] + " " + parts[0]
            if '(' in name:
                split_parts = name.split(' (', 1)
                First = split_parts[0]
                ml = split_parts[-1]
                parts = ml.split(') ',1)
                Middle = parts[0]
                Last = parts[-1]
                fn = First + " " + Last
                mn = Middle + " " + Last
                f_name.append(fn)
                m_name.append(mn)
            else:
                f_name.append(name)
                m_name.append("empty")
        else:
            f_name.append(row)
            m_name.append("empty")
    else:
        f_name.append('empty')
        m_name.append('empty')
        
result['F_name'] = f_name
result['M_name'] = m_name
display(result)
```

```python
from fuzzywuzzy import fuzz
match_score = []
for index, row in result.iterrows():
    if row['M_name'] == "empty":
        title = row['F_name']
        work = row['authorLabel']
        score = fuzz.ratio(title, work)
        match_score.append(score)
    else:
        mt = row['M_name']
        title = row['F_name']
        work = row['authorLabel']
        score1 = fuzz.ratio(title, work)
        score2 = fuzz.ratio(mt, work)
        if score1 > score2:
            match_score.append(score1)
        else:
            match_score.append(score2)
result['name_match'] = match_score
name_match = result[(result['name_match'] >= 80)]
display(result[(result['name_match'] >= 80)])
```

```python
birth_match = result[result['authoryearofbirth'] == result['birthYear']]
display(birth_match)
```

```python
display(result[(result['name_match'] >= 60) & (result['name_match'] < 80)])
```

```python
full_match = pd.concat([birth_match, name_match])
full_match = full_match.drop_duplicates(subset=['id'])
display(full_match)
```

```python
full_match.to_csv("updated_fm.csv")
```

```python
author = full_match[['birthYear','authorLabel','author_QN']]
author = author.drop_duplicates(subset=['author_QN'])
author['birth_after'] = (author['birthYear'] // 10)*10
author['birth_before'] = (author['birthYear'] // 10 + 1) * 10
author['birth_range'] = author['birth_after'].astype(str) + "-" + author['birth_before'].astype(str)
birth_counts = author['birth_range'].value_counts()
print(birth_counts)
```

```python
import matplotlib.pyplot as plt

# Filter birth ranges with counts above 50
filtered_birth_counts = birth_counts[birth_counts > 10]

# Plot a bar chart for birth counts
filtered_birth_counts.plot(kind='bar', edgecolor='black')

# Add labels and title
plt.xlabel('Birth Range')
plt.ylabel('Count')
plt.title('Histogram of Birth Ranges with Counts > 10')

# Show the plot
plt.show()
```
