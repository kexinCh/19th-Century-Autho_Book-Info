```python
import sys
import pandas as pd
from typing import List, Dict
from SPARQLWrapper import SPARQLWrapper, JSON
```

```python
book = pd.read_csv('book_info.csv')
book['book_QN'] = book['work'].str.split('/').str[-1]
book = book.drop_duplicates(subset=['book_QN','genreLabel'])
book = book[book['book_QN'] != "Q43361"]
book_clean = book.drop_duplicates(subset=['book_QN'])
display(book_clean)
```

```python
author = pd.read_csv('author_info.csv')
author['author_QN'] = author['author'].str.split('/').str[-1]
author = author[author['author_QN'] != 'Q34660']
display(author)
```

```python
# pd.set_option('display.max_rows', None)
author_clean = author.drop_duplicates(subset=['author_QN'])
citizen_counts = author_clean['citizenshipLabel'].value_counts()
print(citizen_counts)
```

```python
author_clean['birth_after'] = (author_clean['birthYear'] // 10)*10
author_clean['birth_before'] = (author_clean['birthYear'] // 10 + 1) * 10
author_clean['birth_range'] = author_clean['birth_after'].astype(str) + "-" + author_clean['birth_before'].astype(str)

#We also need author that wrote books in English


display(author_clean)
```

```python
birth_counts = author_clean['birth_range'].value_counts()
print(birth_counts)
```

```python
match = pd.read_csv('newest_fm.csv')
match = match[match['language'] =="['en']"]
display(match)
```

```python
Book_a = pd.merge(match[['book_QN','author_QN']], book_clean[['book_QN','PublicationYear']], on = ['book_QN'])
Book_a = pd.merge(Book_a, author_clean[['author_QN','birthYear']], on = ['author_QN'])
Book_a['publish_age'] = Book_a['PublicationYear'] - Book_a['birthYear']
Book_a['pAge_after'] = (Book_a['publish_age'] // 5)*5
Book_a['pAge_before'] = (Book_a['publish_age'] // 5 + 1) * 5
Book_a['pAge_range'] = Book_a['pAge_after'].astype(str) + "-" + Book_a['pAge_before'].astype(str)
Book_a = Book_a[(Book_a['publish_age'] > 0) & (Book_a['publish_age'] < 100) ]
Book_a_clean = Book_a.drop_duplicates(subset=['book_QN'])
display(Book_a_clean)
```

```python
Eng_books = book_clean[book_clean['book_QN'].isin(Book_a_clean['book_QN'])]
print(Eng_books['genre'].isnull().sum())
display(Eng_books)
```

```python
pAge_counts = Book_a['pAge_range'].value_counts()
print(pAge_counts)
```

```python
book_a = Book_a.drop_duplicates(subset=['book_QN'])
author_book_count = book_a.groupby('author_QN')['book_QN'].count().reset_index(name='book_count')

author_book = author_book_count.sort_values(by='book_count', ascending = False)
# Display the result
display(author_book)
author_book.to_csv('author_book.csv')
```
