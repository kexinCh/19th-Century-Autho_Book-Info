```python
a1 = pd.read_csv('author5000.csv')
a1 = a1.drop_duplicates(subset=['item','work'])
a1 = a1[a1['itemLabel'] != 'J. K. Rowling']
display(a1)
```

```python
a2 = pd.read_csv('author5000-10000.csv')
a2 = a2.drop_duplicates(subset=['item','work'])
a2 = a2[a2['itemLabel'] != 'J. K. Rowling']
display(a2)
```

```python
a3 = pd.read_csv('author10000-16000.csv')
a3 = a3.drop_duplicates(subset=['item','work'])
a3 = a3[a3['itemLabel'] != 'J. K. Rowling']
display(a3)
```

```python
a4 = pd.read_csv('author15000.csv')
a4 = a4.drop_duplicates(subset=['item','work'])
a4 = a4[a4['itemLabel'] != 'J. K. Rowling']
display(a4)
```

```python
a5 = pd.read_csv('author15000.csv')
a5 = a5.drop_duplicates(subset=['item','work'])
a5 = a5[a5['itemLabel'] != 'J. K. Rowling']
display(a5)
```

```python
author = pd.concat([a1,a2,a3,a4,a5])
author = author.drop_duplicates(subset=['item','work'])
author = author.rename(columns={'itemLabel':"F_name","workLabel":"title"})
display(author)
```

```python
metadata = pd.read_csv('update_metadata.csv')
display(metadata)
# matches = pd.merge(metadata, author, on = [''])
```

```python
matches = pd.merge(metadata, author, on = ['F_name','title'])
display(matches)
```

```python
t_matches = pd.merge(metadata, author, on = ['title'])
display(t_matches)
```

```python
t_matches = t_matches.drop_duplicates(subset=['id'])
t_matches['book_QN'] = t_matches['work'].str.split('/').str[-1]
t_matches['author_QN'] = t_matches['item'].str.split('/').str[-1]
t_matches = t_matches[['id','title','language','downloads','book_QN','author_QN','F_name_x']]
t_matches = t_matches.rename(columns={'F_name_x': 'authorLabel'})
display(t_matches)
```

```python
fm = pd.read_csv('updated_fm.csv')
fm = fm[['id','title','language','downloads','book_QN','author_QN','authorLabel']]
display(fm)
```

```python
fm = pd.concat([fm,t_matches])
fm = fm.drop_duplicates(subset=['id'])
display(fm)
fm.to_csv('newest_fm.csv')
```

```python
duplicates = fm.groupby('id')['book_QN'].nunique()
duplicates = duplicates[duplicates > 1]

# Get the rows where the "id" matches and "book_QN" is different
result = fm[fm['id'].isin(duplicates.index)]
result = result.sort_values(by='id')
display(result)
```
